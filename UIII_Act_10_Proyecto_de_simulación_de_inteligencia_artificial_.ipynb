* Este programa simula un Traductor Inteligente con IA diseñado para alumnos de preparatoria. Utiliza una lista de diccionarios para almacenar el conocimiento (palabras y sus traducciones), un ciclo while True para el menú y un ciclo for para recorrer los datos

* Cumpliendo con tus restricciones, el código no utiliza clases, funciones ni bloques de error, e inicia con las estructuras totalmente vacías


```python
# --- SIMULADOR DE IA: TRADUCTOR DINÁMICO ---

# 1. Inicializamos la base de datos como una lista vacía.
# El diccionario para capturar cada entrada también inicia vacío.
lista_traductores = []
nuevo_traductor = {}

# 2. Ciclo 'while True' para mantener el menú interactivo activo [2, 6].
while True:
    print("\n==========================================")
    print("   🤖 IA TRADUCTORA: APRENDIZAJE ACTIVO 🤖")
    print("==========================================")
    print("1. Enseñar nueva palabra a la IA (Agregar)")
    print("2. Traducir una palabra (Buscar con IA)")
    print("3. Mostrar base de datos completa")
    print("4. Apagar IA (Salir)")

    opcion = input("\nSelecciona una opción (1-4): ")

    # 3. Estructuras condicionales 'if-elif' para las decisiones del menú [7, 8].
    if opcion == "1":
        print("\n--- MODO APRENDIZAJE ---")
        palabra_es = input("Escribe la palabra en español: ")
        palabra_en = input("Escribe su traducción al inglés: ")

        # Creamos un diccionario con la nueva traducción [4].
        nuevo_traductor = {"espanol": palabra_es, "ingles": palabra_en}

        # Agregamos el diccionario a la lista (nuestra base de datos) [5].
        lista_traductores.append(nuevo_traductor)
        print(f"¡Logro desbloqueado! La IA ahora conoce: {palabra_es}")

    elif opcion == "2":
        print("\n--- MODO TRADUCCIÓN ---")
        if not lista_traductores:
            print("Error: Mi memoria está vacía. ¡Enséñame algo primero!")
        else:
            buscada = input("¿Qué palabra en español quieres traducir?: ")
            encontrada = False

            # 4. Ciclo 'for' para buscar dentro de la lista de diccionarios [1, 9].
            for registro in lista_traductores:
                # Comparamos la entrada con la clave 'espanol' del diccionario [10].
                if registro["espanol"] == buscada:
                    print(f"IA dice: La traducción de '{buscada}' es '{registro['ingles']}'")
                    encontrada = True

            if encontrada == False:
                print("Lo siento, esa palabra no figura en mi base de datos actual.")

    elif opcion == "3":
        print("\n--- BASE DE DATOS DE LA IA ---")
        # 5. Ciclo 'for' para listar cada diccionario guardado [2].
        if not lista_traductores:
            print("No hay datos registrados aún.")
        else:
            print("Palabras que he aprendido:")
            for item in lista_traductores:
                print(f"• Español: {item['espanol']} | Inglés: {item['ingles']}")

    elif opcion == "4":
        print("\nApagando sistemas... ¡Gracias por entrenarme!")
        # 6. 'break' para salir del ciclo infinito y terminar el programa [6].
        break

    else:
        # Cláusula 'else' para manejar entradas no válidas [7].
        print("\nOpción no válida. Por favor, intenta del 1 al 4.")

print("==========================================")
```
