### ESPECIFICACIONES DE LA CLASE:
- El nombre de la clase sigue la regla de la primera letra en mayúscula y guion bajo: 'Aval_Fiador'.
- Contiene exactamente 5 atributos y 5 funciones principales.

### DEFINICIÓN DE LOS 5 ATRIBUTOS (En mis propias palabras):
1. nombre_avalista: El nombre completo de la persona que sirve como aval, necesario para su identificación legal.
2. dni_identificacion: El documento oficial de identidad (DNI, RUT, RFC) para asegurar que el fiador es localizable.
3. ingresos_mensuales: La cantidad de dinero que gana el avalista al mes para comprobar su capacidad económica de pago.
4. monto_maximo_garantizado: El límite de dinero del préstamo por el cual el aval se hace legalmente responsable.
5. estado_garantia: El estatus actual del compromiso (ej. Pendiente de Firma, Activo, En Reclamación, Liberado).

### DEFINICIÓN DE LAS 5 FUNCIONES (En mis propias palabras):
1. evaluar_solvencia(): Analiza si los ingresos mensuales del aval son suficientes (un 25% o más) respecto al monto garantizado para determinar si es apto.
2. firmar_contrato(): Simula el acto legal donde el aval acepta las condiciones y activa la garantía cambiándola a estado "Activo".
3. actualizar_monto_pendiente(monto_pagado): Resta los pagos que realiza el deudor principal del monto máximo garantizado, reduciendo el riesgo del aval.
4. ejecutar_garantia(monto_impago): Se activa si el deudor principal no paga; cambia el estado a "En Reclamación" y exige el pago al aval.
5. liberar_responsabilidad(): Se ejecuta cuando la deuda se extingue, cambiando el estado a "Liberado" y dejando al aval exento de obligaciones.

```python
class Aval_Fiador:
    def __init__(self, nombre_avalista, dni_identificacion, ingresos_mensuales, monto_maximo_garantizado):
        # [5 Atributos]
        self.nombre_avalista = nombre_avalista
        self.dni_identificacion = dni_identificacion
        self.ingresos_mensuales = ingresos_mensuales
        self.monto_maximo_garantizado = monto_maximo_garantizado
        self.estado_garantia = "Pendiente de Firma"

    # [Función 1] Evaluar solvencia
    def evaluar_solvencia(self):
        # El aval debe ganar al menos el 25% del monto total garantizado
        if self.ingresos_mensuales >= (self.monto_maximo_garantizado * 0.25):
            print(f"\n[✓] Solvencia APROBADA para {self.nombre_avalista}.")
            return True
        else:
            print(f"\n[X] Solvencia RECHAZADA. Los ingresos de ${self.ingresos_mensuales} son insuficientes para respaldar ${self.monto_maximo_garantizado}.")
            return False

    # [Función 2] Firmar contrato
    def firmar_contrato(self):
        if self.estado_garantia == "Pendiente de Firma":
            self.estado_garantia = "Activo"
            print(f"\n[✓] ¡Contrato Firmado! {self.nombre_avalista} ahora es formalmente el aval.")
        else:
            print(f"\n[!] No se puede firmar. El estado actual ya es: {self.estado_garantia}")

    # [Función 3] Actualizar monto por pagos del deudor
    def actualizar_monto_pendiente(self, monto_pagado):
        if self.estado_garantia == "Activo":
            if monto_pagado > self.monto_maximo_garantizado:
                monto_pagado = self.monto_maximo_garantizado

            self.monto_maximo_garantizado -= monto_pagado
            print(f"\n[↓] El deudor pagó ${monto_pagado}.")

            if self.monto_maximo_garantizado <= 0:
                self.liberar_responsabilidad()
            else:
                print(f"-> Responsabilidad restante del aval: ${self.monto_maximo_garantizado}")
        else:
            print(f"\n[!] Operación no permitida. La garantía está en estado: {self.estado_garantia}")

    # [Función 4] Ejecutar garantía por impago
    def ejecutar_garantia(self, monto_impago):
        if self.estado_garantia == "Activo":
            self.estado_garantia = "En Reclamación"
            print(f"\n[¡ALERTA!] El deudor principal cayó en mora.")
            print(f"Se ha notificado a {self.nombre_avalista} (DNI: {self.dni_identificacion}) para cubrir de inmediato: ${monto_impago}")
        else:
            print(f"\n[!] No se puede ejecutar. El estado actual es: {self.estado_garantia}")

    # [Función 5] Liberar responsabilidad
    def liberar_responsabilidad(self):
        self.estado_garantia = "Liberado"
        print(f"\n[★] Enhorabuena. {self.nombre_avalista} ha sido LIBERADO de toda responsabilidad legal.")


# --- FLUJO INTERACTIVO CON INPUT ---
def iniciar_sistema():
    print("=" * 45)
    print("  SISTEMA DE REGISTRO DE AVAL / FIADOR  ")
    print("=" * 45)

    # Captura de datos iniciales para los atributos
    nombre = input("Nombre completo del Aval: ")
    dni = input("Documento de Identidad (DNI/RUT/RFC): ")

    while True:
        try:
            ingresos = float(input("Ingresos mensuales del Aval ($): "))
            monto = float(input("Monto máximo a garantizar ($): "))
            break
        except ValueError:
            print("[X] Por favor, ingresa montos numéricos válidos.\n")

    # Creación del objeto con la clase Aval_Fiador
    aval = Aval_Fiador(nombre, dni, ingresos, monto)

    # Menú de acciones interactivo
    while True:
        print("\n" + "-" * 40)
        print(f" AVAL: {aval.nombre_avalista} | ESTADO: {aval.estado_garantia}")
        print("-" * 40)
        print("1. [Función] Evaluar Solvencia")
        print("2. [Función] Firmar Contrato")
        print("3. [Función] Registrar Pago del Deudor (Actualizar Monto)")
        print("4. [Función] Ejecutar Garantía (Declarar Impago)")
        print("5. [Función] Liberar Responsabilidad")
        print("6. Salir del sistema")

        opcion = input("\nSelecciona una opción (1-6): ")

        if opcion == "1":
            aval.evaluar_solvencia()

        elif opcion == "2":
            aval.firmar_contrato()

        elif opcion == "3":
            try:
                pago = float(input("¿Cuánto pagó el deudor principal?: $"))
                if pago > 0:
                    aval.actualizar_monto_pendiente(pago)
                else:
                    print("[X] El pago debe ser mayor a 0.")
            except ValueError:
                print("[X] Introduce un número válido.")

        elif opcion == "4":
            try:
                impago = float(input("¿Cuál es el monto que el deudor no pagó?: $"))
                if impago > 0:
                    aval.ejecutar_garantia(impago)
                else:
                    print("[X] El monto de impago debe ser mayor a 0.")
            except ValueError:
                print("[X] Introduce un número válido.")

        elif opcion == "5":
            aval.liberar_responsabilidad()

        elif opcion == "6":
            print("\nCerrando el sistema de garantías. ¡Hasta pronto!")
            break
        else:
            print("[X] Opción no válida. Intenta de nuevo.")

# Ejecutar el programa interactivo
if __name__ == "__main__":
    iniciar_sistema()
```
