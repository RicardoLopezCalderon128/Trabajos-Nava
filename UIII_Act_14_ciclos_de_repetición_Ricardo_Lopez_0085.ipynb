* Este es un programa en Python que simula una cuenta bancaria interactiva. Utiliza un bucle while para permitir operaciones constantes y estructuras if-elif-else para la toma de decisiones, cumpliendo estrictamente con las restricciones de no usar funciones, listas, diccionarios ni programación orientada a objetos


```python
# --- SIMULADOR DE CUENTA BANCARIA ---

# 1. Se inicializa la variable del saldo en cero [2, 4].
saldo = 0.0

# 2. Se define una variable para controlar el estado del bucle [3].
continuar = "si"

# 3. El ciclo 'while' se ejecuta mientras el usuario quiera seguir y el saldo no sea negativo [1, 5, 6].
while continuar == "si" and saldo >= 0:
    print(f"\n--- Saldo actual: ${saldo} ---")
    print("Opciones: (1) Depósito, (2) Retiro, (3) Salir")

    # 4. Se utiliza input() para capturar la opción del usuario [7, 8].
    opcion = input("Seleccione una operación: ")

    # 5. Estructura condicional para manejar las operaciones [7, 9].
    if opcion == "1":
        # 6. Se solicita el monto y se convierte a decimal (float) para cálculos [7, 10].
        deposito = float(input("Ingrese el monto a depositar: "))
        # 7. Operador aritmético para sumar el depósito al saldo [4, 11].
        saldo = saldo + deposito
        print("Depósito realizado con éxito.")

    elif opcion == "2":
        retiro = float(input("Ingrese el monto a retirar: "))
        # 8. Operador aritmético para restar el retiro del saldo [4].
        saldo = saldo - retiro
        print("Retiro procesado.")

        # 9. Verificación inmediata de saldo negativo para cerrar la cuenta [7, 12].
        if saldo < 0:
            print("¡ALERTA!: Su saldo es negativo. La cuenta se bloqueará.")
            # El bucle terminará automáticamente en la siguiente evaluación de la condición.

    elif opcion == "3":
        # 10. Valor centinela para que el usuario decida salir voluntariamente [2, 13].
        continuar = "no"
        print("Saliendo del sistema bancario.")

    else:
        print("Opción no válida, intente de nuevo.")

# 11. Mensaje final al terminar el ciclo [14].
print("\n--- Proceso finalizado. Gracias por usar nuestro servicio ---")
```
