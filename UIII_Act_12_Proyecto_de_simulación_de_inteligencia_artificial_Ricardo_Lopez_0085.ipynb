* Este programa utiliza la Programación Orientada a Objetos (POO) para crear un sistema interactivo de gestión de pilotos. Una clase funciona como una plantilla o "molde" que define los datos (atributos) y los comportamientos (métodos) de un objeto

* En este ejemplo, la clase Piloto encapsula la información de cada profesional, y utilizamos una lista para almacenar las instancias (objetos individuales) creadas a partir de dicha clase

```python
# --- SISTEMA DE GESTIÓN AERONÁUTICA CON POO ---

class Piloto:
    def __init__(self, nombre, horas, aeronave, aerolinea, rango, licencia, certificado):
        """Método constructor para inicializar los atributos del piloto [5, 6]"""
        self.nombre = nombre
        self.horas = horas
        self.aeronave = aeronave
        self.aerolinea = aerolinea
        self.rango = rango
        self.licencia = licencia
        self.certificado = certificado

    def analizar_ia(self):
        """Simulación de IA: Realiza un análisis lógico basado en los datos del objeto"""
        if self.certificado.lower() == "no":
            return "❌ NO APTO: Certificado médico vencido."
        elif self.horas > 1500 and self.rango.lower() == "oficial":
            return "⭐ PROMOCIÓN: Candidato para Capitán."
        elif self.horas < 200:
            return "⚠️ ADVERTENCIA: Horas insuficientes para vuelos comerciales."
        else:
            return "✅ APTO: Estatus operativo normal."

# Lista global para almacenar los objetos de la clase Piloto [7]
bd_pilotos = []

def agregar_piloto():
    print("\n--- Registro de Nuevo Piloto ---")
    nombre = input("Nombre: ")
    horas = int(input("Horas de vuelo: "))
    aeronave = input("Tipo de aeronave: ")
    aerolinea = input("Aerolínea: ")
    rango = input("Rango (Capitán/Oficial): ")
    licencia = input("¿Licencia Internacional? (Si/No): ")
    certificado = input("¿Certificado Médico? (Si/No): ")

    # Se crea el objeto (instanciación) y se agrega a la lista [8, 9]
    nuevo = Piloto(nombre, horas, aeronave, aerolinea, rango, licencia, certificado)
    bd_pilotos.append(nuevo)
    print(f"¡Piloto {nombre} registrado con éxito!")

def listar_pilotos():
    if not bd_pilotos:
        print("\nLa base de datos está vacía.")
    else:
        print("\n" + "="*125)
        # Encabezados alineados usando f-strings para una tabla profesional
        print(f"{'NOMBRE':<20} | {'HORAS':<7} | {'AERONAVE':<15} | {'AEROLÍNEA':<15} | {'RANGO':<10} | {'IA ESTATUS'}")
        print("-" * 125)
        for p in bd_pilotos:
            # Se accede a los atributos mediante la notación de punto (objeto.atributo) [10]
            print(f"{p.nombre:<20} | {p.horas:<7} | {p.aeronave:<15} | {p.aerolinea:<15} | {p.rango:<10} | {p.analizar_ia()}")
        print("="*125)

def actualizar_pilotos():
    nombre_buscado = input("\nNombre del piloto a actualizar: ")
    encontrado = False
    for p in bd_pilotos:
        if p.nombre.lower() == nombre_buscado.lower():
            print(f"Modificando datos de {p.nombre}...")
            p.horas = int(input("Nuevas horas de vuelo: "))
            p.rango = input("Nuevo rango: ")
            p.certificado = input("¿Certificado actualizado? (Si/No): ")
            encontrado = True
            print("Datos actualizados.")
            break
    if not encontrado:
        print("Piloto no encontrado.")

def borrar_piloto():
    nombre_buscado = input("\nNombre del piloto a eliminar: ")
    for p in bd_pilotos:
        if p.nombre.lower() == nombre_buscado.lower():
            bd_pilotos.remove(p)
            print(f"Piloto {nombre_buscado} eliminado del sistema.")
            return
    print("No se encontró el piloto.")

# --- MENÚ INTERACTIVO PRINCIPAL ---
while True:
    print("\n========= MENÚ AERONÁUTICO IA =========")
    print("1. Agregar Piloto")
    print("2. Listar Pilotos (Tabla)")
    print("3. Actualizar Piloto")
    print("4. Borrar Piloto")
    print("5. Salir")

    opcion = input("Seleccione una opción (1-5): ")

    if opcion == "1":
        agregar_piloto()
    elif opcion == "2":
        listar_pilotos()
    elif opcion == "3":
        actualizar_pilotos()
    elif opcion == "4":
        borrar_piloto()
    elif opcion == "5":
        print("Saliendo del simulador... ¡Buen vuelo!")
        break
    else:
        print("Opción inválida.")
```
