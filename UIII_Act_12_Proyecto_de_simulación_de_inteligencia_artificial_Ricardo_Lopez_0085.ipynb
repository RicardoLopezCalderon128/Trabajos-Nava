* Este programa utiliza la Programación Orientada a Objetos (POO) para crear un sistema interactivo de gestión de pilotos. Una clase funciona como una plantilla o "molde" que define los datos (atributos) y los comportamientos (métodos) de un objeto

* En este ejemplo, la clase Piloto encapsula la información de cada profesional, y utilizamos una lista para almacenar las instancias (objetos individuales) creadas a partir de dicha clase

```python
# --- SIMULADOR DE IA AERONÁUTICA: GESTIÓN DE PILOTOS ---

class Piloto:
    def __init__(self, nombre, horas, aeronave, aerolinea, rango, licencia, certificado):
        """Constructor: Inicializa los atributos del piloto [8-10]"""
        self.nombre = nombre
        self.horas = horas
        self.aeronave = aeronave
        self.aerolinea = aerolinea
        self.rango = rango
        self.licencia = licencia
        self.certificado = certificado

    def analizar_ia(self):
        """Simulación de IA: Lógica de decisión basada en atributos [11, 12]"""
        if self.certificado.lower() == "no":
            return "🔴 BLOQUEADO: Certificado médico vencido."
        elif self.horas > 5000:
            return "🟢 EXPERTO: Apto para rutas transoceánicas."
        elif self.horas < 200:
            return "🟡 CADETE: Requiere supervisión en cabina."
        else:
            return "🔵 OPERATIVO: Estatus estándar verificado."

    def __init__(self):
        # Lista vacía para almacenar los objetos Piloto [13, 14]
        self.lista_pilotos = []

    def agregar_piloto(self):
        """Solicita datos y crea una nueva instancia de Piloto [15, 16]"""
        print("\n--- Registro de Nuevo Piloto ---")
        nom = input("Nombre completo: ")
        hrs = int(input("Horas de vuelo acumuladas: "))
        aero = input("Tipo de aeronave: ")
        linea = input("Aerolínea: ")
        ran = input("Rango (Capitán/Oficial): ")
        lic = input("¿Licencia internacional vigente? (Si/No): ")
        cert = input("¿Certificado médico vigente? (Si/No): ")

        # Se crea el objeto y se agrega a la lista [17, 18]
        nuevo_piloto = Piloto(nom, hrs, aero, linea, ran, lic, cert)
        self.lista_pilotos.append(nuevo_piloto)
        print(f"IA: Piloto {nom} registrado en la base de datos.")

    def listar_pilotos(self):
        """Muestra los pilotos en una tabla alineada con f-strings [19, 20]"""
        if not self.lista_pilotos:
            print("\nIA: No hay pilotos registrados actualmente.")
        else:
            print("\n" + "="*120)
            # Encabezados con distribución de espacios
            print(f"{'PILOTO':<20} | {'HORAS':<6} | {'AERONAVE':<15} | {'RANGO':<10} | {'IA: ANÁLISIS DE ESTATUS'}")
            print("-" * 120)
            for p in self.lista_pilotos:
                # Acceso a atributos y ejecución de la función de IA [5, 21]
                print(f"{p.nombre:<20} | {p.horas:<6} | {p.aeronave:<15} | {p.rango:<10} | {p.analizar_ia()}")
            print("="*120)

    def actualizar_pilotos(self):
        """Busca un piloto y modifica sus atributos [22, 23]"""
        buscado = input("\nNombre del piloto a actualizar: ")
        encontrado = False
        for p in self.lista_pilotos:
            if p.nombre.lower() == buscado.lower():
                print(f"Modificando datos de {p.nombre}...")
                p.horas = int(input("Nuevas horas de vuelo: "))
                p.rango = input("Nuevo rango: ")
                p.certificado = input("¿Certificado médico actualizado? (Si/No): ")
                encontrado = True
                print("IA: Información actualizada correctamente.")
        if not encontrado:
            print("IA: El piloto no existe en mis registros.")

    def borrar_piloto(self):
        """Elimina un objeto de la lista [24, 25]"""
        buscado = input("\nNombre del piloto a eliminar: ")
        for p in self.lista_pilotos:
            if p.nombre.lower() == buscado.lower():
                self.lista_pilotos.remove(p)
                print(f"IA: El registro de {buscado} ha sido borrado.")
                return
        print("IA: No se encontró al piloto para eliminar.")

# --- MENÚ INTERACTIVO PRINCIPAL [26, 27] ---
gestor = SistemaIA()

while True:
    print("\n----------- MENÚ DE CONTROL IA -----------")
    print("1. Agregar Piloto")
    print("2. Listar Pilotos (Tabla)")
    print("3. Actualizar Piloto")
    print("4. Borrar Piloto")
    print("5. Salir")

    opc = input("Seleccione una opción (1-5): ")

    if opc == "1":
        gestor.agregar_piloto()
    elif opc == "2":
        gestor.listar_pilotos()
    elif opc == "3":
        gestor.actualizar_pilotos()
    elif opc == "4":
        gestor.borrar_piloto()
    elif opc == "5":
        print("IA: Desconectando sistema... ¡Buen vuelo!")
        break
    else:
        print("Opción no válida.")
```
