#### traductor
* Este programa simula un Traductor Inteligente que "aprende" nuevas palabras. Utiliza dos listas para almacenar el conocimiento (una para español y otra para inglés) y se controla totalmente mediante un menú interactivo con el ciclo while.
* Cumpliendo con tus requisitos, el código no utiliza funciones, ni clases (POO), ni ciclos for, ni bloques try. El recorrido de las listas se realiza manualmente mediante índices y el ciclo while.

```python
# --- SIMULADOR DE IA: TRADUCTOR DINÁMICO ---

# Inicializamos las listas de conocimiento vacías
# En Python, las listas son colecciones que pueden crecer [1, 2]
conocimiento_esp = []
conocimiento_eng = []

opcion = ""

# Iniciamos el ciclo principal para el menú interactivo [3, 4]
while opcion != "4":
    print("\n==========================================")
    print("   🤖 IA TRADUCTORA: APRENDIZAJE ACTIVO 🤖")
    print("==========================================")
    print("1. Enseñar nueva palabra a la IA")
    print("2. Traducir palabra")
    print("3. Ver base de datos de la IA")
    print("4. Apagar IA")

    opcion = input("\nSelecciona una opción (1-4): ")

    # Estructura condicional para procesar la opción del usuario [5, 6]
    if opcion == "1":
        print("\n--- MODO APRENDIZAJE ---")
        palabra_es = input("Palabra en español: ")
        palabra_en = input("Traducción en inglés: ")

        # Guardamos los datos en las listas usando append [2]
        conocimiento_esp.append(palabra_es)
        conocimiento_eng.append(palabra_en)
        print("¡Entendido! He guardado la palabra en mi memoria.")

    elif opcion == "2":
        print("\n--- MODO TRADUCCIÓN ---")
        buscada = input("¿Qué palabra quieres traducir?: ")

        # Para buscar sin usar for, usamos un while con un índice contador [7]
        indice = 0
        encontrada = False
        largo_lista = len(conocimiento_esp) # Obtenemos cuántas palabras sé [8]

        while indice < largo_lista and encontrada == False:
            # Comparamos la entrada con lo almacenado [9]
            if conocimiento_esp[indice] == buscada:
                print(f"La traducción es: {conocimiento_eng[indice]}")
                encontrada = True
            indice = indice + 1 # Incrementamos el contador para pasar al siguiente

        if encontrada == False:
            print("Lo siento, esa palabra no está en mi base de datos.")

    elif opcion == "3":
        print("\n--- BASE DE DATOS DE LA IA ---")
        # Recorremos la lista usando un contador manual [7]
        indice = 0
        largo_lista = len(conocimiento_esp)

        if largo_lista == 0:
            print("Mi memoria está vacía. ¡Enséñame algo primero!")
        else:
            while indice < largo_lista:
                # Accedemos a los elementos por su posición de índice [2]
                print(f"{indice + 1}. {conocimiento_esp[indice]} -> {conocimiento_eng[indice]}")
                indice = indice + 1

    elif opcion == "4":
        print("\nIA finalizada. ¡Hasta luego, estudiante!")

    else:
        print("\nOpción no válida. Por favor, intenta del 1 al 4.")

print("==========================================")
```
