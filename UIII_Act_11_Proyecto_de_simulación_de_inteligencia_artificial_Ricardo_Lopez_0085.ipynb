* Este programa simula un Sistema de Gestión y Asistente Lógico de Pilotos. Utiliza una lista de diccionarios para almacenar la información de cada piloto, donde cada "ficha" es un diccionario con sus datos técnicos

* El programa es interactivo, utiliza funciones para organizar el código
 y presenta un "Asistente de IA" que evalúa automáticamente el estatus del piloto basándose en sus horas de vuelo y certificados

```python
# --- SISTEMA DE GESTIÓN AERONÁUTICA: PILOTOS ---

# Inicializamos la lista de diccionarios vacía
# Esta lista actuará como nuestra base de datos dinámica [2, 7]
lista_pilotos = []

def agregar_piloto():
    """Captura los datos del piloto y los guarda en un diccionario [8, 9]."""
    print("\n--- REGISTRO DE NUEVO PILOTO ---")

    # Creamos el diccionario vacío para el piloto actual
    piloto = {}

    # Entrada de datos (Key: Value)
    # Se utiliza input() para capturar los valores [10, 11]
    nombre = input("Nombre del Piloto: ")
    horas = int(input("Horas de vuelo acumuladas: "))
    aeronave = input("Tipo de aeronave (ej. Boeing 737, Airbus A320): ")
    aerolinea = input("Aerolínea: ")
    rango = input("Rango (Capitán / Oficial): ")
    licencia = input("¿Licencia Internacional vigente? (Si/No): ")
    certificado = input("¿Certificado Médico vigente? (Si/No): ")

    # Asignamos las claves y valores al diccionario [2, 12]
    piloto["nombre"] = nombre
    piloto["horas"] = horas
    piloto["aeronave"] = aeronave
    piloto["aerolinea"] = aerolinea
    piloto["rango"] = rango
    piloto["licencia"] = licencia
    piloto["certificado"] = certificado

    # Agregamos el diccionario a nuestra lista global [7, 12]
    lista_pilotos.append(piloto)
    print(f"\n¡Piloto {nombre} registrado exitosamente!")

def listar_pilotos():
    """Muestra los pilotos en una tabla bien distribuida [13, 14]."""
    if not lista_pilotos:
        print("\n[!] La base de datos está vacía.")
    else:
        print("\n" + "="*115)
        # Encabezados de la tabla alineados con f-strings [13]
        print(f"{'NOMBRE':<20} | {'HORAS':<7} | {'AERONAVE':<15} | {'AEROLÍNEA':<15} | {'RANGO':<10} | {'LIC. INT.':<8} | {'MÉDICO':<6}")
        print("-" * 115)

        # Ciclo for para recorrer la lista de diccionarios [15, 16]
        for p in lista_pilotos:
            print(f"{p['nombre']:<20} | {p['horas']:<7} | {p['aeronave']:<15} | {p['aerolinea']:<15} | {p['rango']:<10} | {p['licencia']:<8} | {p['certificado']:<6}")
        print("="*115)

def asistente_ia():
    """Simula una IA que analiza el estatus de los pilotos [5, 17]."""
    print("\n--- ASISTENTE LÓGICO DE IA (ANÁLISIS DE ESTATUS) ---")
    busqueda = input("Nombre del piloto a analizar: ")
    encontrado = False

    for p in lista_pilotos:
        if p["nombre"].lower() == busqueda.lower():
            encontrado = True
            print(f"\n> IA Analizando a: {p['nombre']}...")

            # Lógica de decisión de la IA (if, elif, else) [5, 6]
            if p["certificado"].lower() == "no":
                print("ESTADO: ¡ALERTA! El piloto no puede volar por Certificado Médico vencido.")
            elif p["horas"] > 1500 and p["rango"].lower() == "oficial":
                print("ESTADO: Candidato apto para promoción a Rango de Capitán.")
            elif p["horas"] < 200:
                print("ESTADO: Piloto en fase de entrenamiento (Horas insuficientes para vuelos comerciales).")
            elif p["licencia"].lower() == "si" and p["horas"] >= 500:
                print("ESTADO: Piloto calificado para Rutas Internacionales.")
            else:
                print("ESTADO: Estatus normal. Operaciones estándar permitidas.")

    if not encontrado:
        print("IA: No se encontró ningún piloto con ese nombre.")

# --- MENÚ PRINCIPAL ---
# El ciclo while True mantiene el programa activo hasta que el usuario decida salir [18, 19]
while True:
    print("\n" + "*"*35)
    print("  SISTEMA AERONÁUTICO INTERACTIVO  ")
    print("*"*35)
    print("1. Registrar Piloto")
    print("2. Ver Tabla de Pilotos")
    print("3. Consultar Asistente IA")
    print("4. Salir")

    opcion = input("\nSeleccione una opción: ")

    # Estructuras selectivas para el menú [5, 20]
    if opcion == "1":
        agregar_piloto()
    elif opcion == "2":
        listar_pilotos()
    elif opcion == "3":
        asistente_ia()
    elif opcion == "4":
        print("\nCerrando sistema... ¡Buen vuelo!")
        break # Finaliza el ciclo while [19]
    else:
        print("\n[!] Opción no válida, intente de nuevo.")
```
